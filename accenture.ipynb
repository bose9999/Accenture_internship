```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Content = pd.read_csv('/content/Content.csv')
ReactionTypes = pd.read_csv('/content/ReactionTypes.csv')
Reactions = pd.read_csv('/content/Reactions.csv')
```

```python
Content.head()
```

```python
Content=Content.drop(['Unnamed: 0'],axis=1)
```

```python
Content.head()
```

```python
Content=Content.drop(['URL'],axis=1)
```

```python
Content.head()
```

```python
Content['Category'].astype(str)
```

```python
Content = Content.drop(['User ID'],axis=1)
```

```python
Content.head()
```

```python
Content= Content.rename(columns={"Type": 'Content Type'})
```

```python
Content.head()
```

```python
def turn(x):
    x =x.replace('"', "")
    return x


Content["Category"]=Content["Category"].apply(lambda x : turn(x) )
```

```python
Content['Category']
```

```python
ReactionTypes.head()
```

```python
ReactionTypes=ReactionTypes.drop(['Unnamed: 0'],axis=1)
```

```python
ReactionTypes= ReactionTypes.rename(columns={"Type": 'Reaction Type'})
```

```python
ReactionTypes.head()
```

```python
Reactions.head()
```

```python
Reactions= Reactions.rename(columns={"Type": 'Reaction Type'})
```

```python
Reactions=Reactions.drop(['Unnamed: 0'],axis=1)
```

```python
Reactions.head()
```

```python
Reactions=Reactions.drop(['User ID'],axis=1)
```

```python
Reactions.head()
```

```python
Reactions = Reactions.merge(Content[["Content ID",'Category']], left_on="Content ID", right_on="Content ID",how = 'inner')
```

```python
Reactions.head()
```

```python
Reactions = Reactions.merge(ReactionTypes[["Reaction Type",'Score']], left_on="Reaction Type", right_on="Reaction Type",how = 'inner')
```

```python
Reactions.head()
```

```python
Reactions
```

```python
Merged_dataset = Reactions.copy()
```

```python
Merged_dataset
```

```python
len(Merged_dataset["Category"] == "veganism")
```

```python
Merged_dataset["Category"].max()
```

```python
Reactions = Reactions.groupby("Category")["Score"].sum()
```

```python
Reactions =Reactions.sort_values(ascending=False)
```

```python
Reactions.to_csv('merged_data.csv')
```

```python
from datetime import datetime

Merged_dataset['Datetime'] = pd.to_datetime(Merged_dataset['Datetime'] ,infer_datetime_format=True)
```

```python
Merged_dataset["Month"] = Merged_dataset["Datetime"].dt.month
```

```python
Merged_dataset["MonthName"] = Merged_dataset["Datetime"].dt.strftime("%B")
```

```python
Merged_dataset
```

```python
d1 = Merged_dataset.copy()
```

```python
d1["Count"] = d1.groupby('Month')['Month'].transform('count')
```

```python
d1 =d1.drop_duplicates(subset=["Month"])
```

```python
d1.reset_index()
```

```python
fig = plt.figure(figsize=(10, 6), dpi=200)
ax = plt.subplot()
x = d1["Month"]
y = d1["Count"]
ax.set_xticks(np.arange(1, 13))  # Set the x-ticks from 1 to 12
ax.set_xticklabels(d1["MonthName"], rotation=45)  # Set the month names as x-tick labels with rotation

plt.title("Posts per Month")
plt.xlabel("MonthName")
plt.ylabel("Posts")

colors = ["orange" if i != 5 else "green" for i in x]
plt.bar(x, y, color=colors)

plt.show()
```
